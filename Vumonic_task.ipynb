## Parsing Data from email receipts and storing it in a CSV file

```python
import os
import pandas as pd
```

### lisitng out all the files in the driectory

```python
folder_path=r'C:\Users\HP\Desktop\mic_research\email_receipts'
files=os.listdir(folder_path)
```

### Using mindee api to scrape data from receipts

```python
from mindee import Client
  
my_client = Client().config_custom_doc(
    document_type="email_receipt",
    singular_name="custom_doc",
    plural_name="custom_docs",
    account_name="gsprabhu01",
    api_key="461b219887164a4a9dfd7a8367384306"
)
```

#### creating the dataframe

```python
columns=['Order_id','Product_name','total_cost','address']
dt=pd.DataFrame(columns=columns)
```

## code to scrape data and append to dataframe

```python
for i in range(7):
    
    #parsing through each receipt in the folder 
    my_doc = my_client.doc_from_path(folder_path+"\\"+files[i])
    parsed_doc = my_doc.parse("email_receipt")
    
    #storing feature data
    total_incl = parsed_doc.custom_doc.total_cost
    address= parsed_doc.custom_doc.address
    product_name= parsed_doc.custom_doc.product_name
    order_id= parsed_doc.custom_doc.order_id

    #creating list for multi word features such as address and product name
    l=[]
    for i in address['values']:
        l.append(i['content'])
    l=" ".join(l)    

    k=[]
    for i in product_name['values']:
        k.append(i['content'])
    k=" ".join(k)    

    #appending to the dataframe
    dt=dt.append({'Order_id':order_id['values'][0]['content'],'Product_name':k,'total_cost':total_incl['values'][0]['content'],'address':l},ignore_index=True)
```
